```python
import cv2
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/Colab_Notebooks/Deep_Learning/Progetto/Input/Input_resize/0.png"

img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

nuove_dimensioni = (128, 128)
img_ridimensionata = cv2.resize(img, nuove_dimensioni, interpolation=cv2.INTER_LINEAR)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Immagine Originale')

plt.subplot(1, 2, 2)
plt.imshow(img_ridimensionata, cmap='gray')
plt.title('Immagine Ridimensionata (Bilineare)')

plt.show()
```

```python
from pathlib import Path

path = Path("/content/drive/MyDrive/Colab_Notebooks/Deep_Learning/Progetto/Input/Input_resize/")
for p in path.iterdir():
  number = str(p).split('/')[-1].split('.')[0]

  img = cv2.imread(str(p), cv2.IMREAD_GRAYSCALE)
  nuove_dimensioni = (128, 128)
  img_ridimensionata = cv2.resize(img, nuove_dimensioni, interpolation=cv2.INTER_LINEAR)

  plt.imsave(f'/content/drive/MyDrive/Colab_Notebooks/Deep_Learning/Progetto/Input/Input_model/mat{number}.png',img_ridimensionata, cmap='gray')
```
